# Applying linear regression to predict the CTC

```python
from sklearn.linear_model import LinearRegression
```

```python
x_train=df.drop("CTC",axis=1)
y_train=df["CTC"]
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
df_test=pd.read_excel(r"C:\Users\kingr\Downloads\Test_data_file.xlsx")
```

```python
df_test=pd.get_dummies(df_test,drop_first=True)
df_test
```

```python
X_test=df_test.drop('Actual CTC',axis=1)
y_test=df_test['Actual CTC']
```

```python
pred = model.predict(X_test)
```

# Evaluating the results of the analysis

```python
from sklearn.metrics import mean_squared_error
print("Mean squared error =", mean_squared_error(y_test,pred))
```
